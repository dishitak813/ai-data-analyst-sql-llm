# Task 2: Data Cleaning & Validation

## Step 1: Identifying and Removing Duplicate Rows

In this step, we identify duplicate records in the Shopify dataset.

Duplicate rows can distort sales metrics and lead to incorrect analysis, so they are removed to maintain data integrity.

```python
import pandas as pd

# Load dataset
df = pd.read_csv("../data/Raw_Shopify_Sales.csv")

# Initial shape
print("Initial shape: ", df.shape)

# Preview data
df.head()
```

```python
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)
```

### Detection Method

We used the `.duplicated()` function in pandas to detect rows that are exact duplicates across all columns.

This method flags rows that have identical values across all fields.

```python
df = df.drop_duplicates()
```

```python
duplicate_count_after = df.duplicated().sum()
print("Duplicates after removal:", duplicate_count_after)
```

### Result

Duplicate rows were successfully removed, ensuring that each record in the dataset is unique and reliable for further analysis.

## Step 2: Standardizing Date Format

In this step, we convert all date-related columns into proper datetime format.

The dataset contains multiple date columns:
- Date (DD-MM-YYYY format)
- Transaction Timestamp (ISO format)
- Order Created At (ISO format)
- Order Updated At (ISO format)

Since formats are different, we handle them separately.

```python
df.columns
```

```python
df[['Date', 'Transaction Timestamp', 'Order Created At', 'Order Updated At']].head()
```

### Converting Date Column

The 'Date' column follows DD-MM-YYYY format, so we use `dayfirst=True`.

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

df['Date'].head()
```

### Converting Timestamp Columns

The remaining columns are already in ISO format (YYYY-MM-DDTHH:MM:SSZ),
so we convert them without using `dayfirst=True`.

```python
timestamp_cols = ['Transaction Timestamp', 'Order Created At', 'Order Updated At']

for col in timestamp_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

df[timestamp_cols].head()
```

### Checking Data Types

We verify that all date columns are successfully converted to datetime format.

```python
df[['Date'] + timestamp_cols].dtypes
```

### Checking Invalid Dates

We count how many values could not be converted and became NaT.

```python
print("Date nulls:", df['Date'].isnull().sum())

for col in timestamp_cols:
    print(col, "nulls:", df[col].isnull().sum())
```

### Validating Date Logic

We check logical consistency:

- Order Created At should be before or equal to Order Updated At
- Transaction Timestamp should be after Order Created At

Invalid rows are flagged.

```python
df['invalid_order_dates'] = (
    (df['Order Created At'].notnull()) &
    (df['Order Updated At'].notnull()) &
    (df['Order Created At'] > df['Order Updated At'])
)

df['invalid_transaction_time'] = (
    (df['Transaction Timestamp'].notnull()) &
    (df['Order Created At'].notnull()) &
    (df['Transaction Timestamp'] < df['Order Created At'])
)

print("Invalid order dates:", df['invalid_order_dates'].sum())
print("Invalid transaction times:", df['invalid_transaction_time'].sum())
```

## Step 3: Handling Missing Values

In this step, we analyze missing values across the dataset and apply a clear, justified strategy.

Instead of blindly dropping rows, we:
- Identify missing values
- Understand their impact
- Apply appropriate imputation or leave them as is where necessary

```python
df.isnull().sum().sort_values(ascending=False).head(15)
```

### Observing Missing Data

We inspect which columns contain missing values and decide strategies based on column importance.

### Strategy for Handling Missing Values

We apply different strategies based on data type and business logic:

1. Date columns → keep NaT (missing dates may be meaningful)
2. Numeric columns → fill with 0 where appropriate (e.g., sales, revenue)
3. Categorical columns → fill with 'Unknown'
4. IDs → do not modify (important identifiers)

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

numeric_cols
```

```python
df[numeric_cols] = df[numeric_cols].fillna(0)
```

### Handling Categorical Columns

Missing categorical values are replaced with 'Unknown' to maintain consistency.

```python
categorical_cols = df.select_dtypes(include=['object']).columns

df[categorical_cols] = df[categorical_cols].fillna('Unknown')
```

### Keeping Date Columns Unchanged

Date columns are left as NaT because missing dates may indicate incomplete transactions.

```python
df.isnull().sum().sort_values(ascending=False).head(10)
```

### Observations

- Numeric values were filled with 0 to avoid calculation issues
- Categorical values were filled with 'Unknown'
- Date columns were preserved as NaT to retain missing information
- This approach ensures minimal data loss while maintaining analytical usability

## Step 4: Recalculating Metrics (CTR, CPC, CPM, ROI)

In this step, we compute key performance metrics from available columns.

Since original metrics are not present in the dataset, we:
- Derive them using source columns
- Identify invalid or unrealistic values

### Note on Metric Availability

The Shopify dataset does not contain required columns such as impressions, clicks, or ad spend.

Therefore:
- CTR (Click Through Rate) cannot be calculated
- CPC (Cost Per Click) cannot be calculated
- CPM (Cost Per 1000 Impressions) cannot be calculated

However, we can calculate ROI using available revenue-related fields.

### Calculating ROI (Return on Investment)

ROI is calculated as:

ROI = (Net Sales - Costs) / Costs

Since cost data is not explicitly available, we approximate using:
- Net Sales (INR)
- Discounts (INR) as a proxy for cost impact

```python
df['ROI_calculated'] = (
    df['Net Sales (INR)'] - df['Discounts (INR)']
) / df['Discounts (INR)']
```

```python
df['ROI_calculated'] = df['ROI_calculated'].replace([float('inf'), -float('inf')], 0)
df['ROI_calculated'] = df['ROI_calculated'].fillna(0)
```

### Handling Invalid ROI Values

- Division by zero may produce infinite values
- These are replaced with 0 for stability

```python
df['ROI_flag'] = (df['ROI_calculated'] > 10) | (df['ROI_calculated'] < -1)
```

### Flagging Suspicious Values

We flag rows where ROI is:
- Extremely high (>10)
- Extremely low (< -1)

These may indicate data quality issues or anomalies.

```python
df['ROI_flag'].sum()
```

```python
df[df['ROI_flag'] == True][
    ['Net Sales (INR)', 'Discounts (INR)', 'ROI_calculated']
].head(10)
```

## Step 5: Normalising String Columns

In this step, we clean and standardise all string-based columns.

This includes:
- Converting text to consistent format
- Removing extra spaces
- Handling missing values
- Standardising naming conventions

```python
string_cols = df.select_dtypes(include='object').columns
string_cols
```

### Cleaning Strategy

We apply the following transformations:
- Strip leading/trailing spaces
- Convert text to a consistent format (Title Case)
- Preserve null values correctly

```python
for col in string_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()

df['Data Source name'].head()
```

```python
for col in string_cols:
    df[col] = df[col].replace(['NAN', 'nan', 'none'], None)

df['Product Tags'].head(20)
```
